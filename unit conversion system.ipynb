```python

#Distance conversion functions

                              ########  LightYear___TO___KiloMeters  ###########
    
def cups_to_ounces():
    print('### YOU CAN ENTER VALUES FROM 1 TO 10) ###')
    cups = float(input("Enter Volume in Cups :: "))
    if cups>0 and cups<11:
        ounces = cups * 8
        print('{0} cups is equal to {1} ounces. '.format(cups, ounces))
       
    else:
        print("invalid input")
        cups_to_ounces()
 #Volume conversion functions

                    ########  Cups___TO___Ounce  ###########
        
    
def lightYear_kilometer():
    print('### YOU CAN ENTER VALUES FROM 1 TO 10) ###')
    lightYear = float(input("Enter a distance in light-years: "))
    if lightYear>0 and lightYear<11:
        kilometer = lightYear * (9.4607304725808*(10**12))
        print('{0} light-years is equal to {1:4f} kilometers. '.format(lightYear, kilometer))
    else:
        print("Invalid input:")
        lightYear_kilometer()






#Temperature conversion functions

                    ########  FAHRENHEIT___TO___CELSIUS  ###########

def fahrenheit_to_celsius():
    print('### YOU CAN ENTER VALUES FROM 1 TO 10) ###')
    fahrenheit = float(input("Enter a temperature value in degrees Fahrenheit :: "))
    if fahrenheit>0 and  fahrenheit<11:
        celsius = (fahrenheit - 32)*5/9
        print("{0} degrees Fahrenheit is equal to {1} degrees Celsius. ".format(fahrenheit, celsius))
    else:
        print("Invalid input:")
        fahrenheit_to_celsius()

                 
########  MAIN FUNCTION  ###########


def main():
    selection = 0

    while selection  !=4:
        print('\n')
        print('### Well Come to My programe ###')
        print('Press 1 for distance Conversion ')
        print('Press 2 for mass conversion')
        print('Press 3 for temprature conversion')
        print('Press 4 to Exit')
        print('\n\n')

        selection = int(input("Select number from 1 to 4 to perform Task :: "))
        if selection > 4 or selection < 1:
            print('Wrong Input !!. Please enter Mentioned Numbers :)')
        else:   
            if selection == 1:
                lightYear_kilometer()
            elif selection == 2:
                cups_to_ounces()
            elif selection == 3:
                fahrenheit_to_celsius()
            else:
                break
    



if __name__ == '__main__':
    main()
    
    print("Thanks for Use")

   
```
